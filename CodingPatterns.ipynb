# Sliding Window

Problems in Sliding Window

Sliding Window has two types:
1. Fixed Window
2. Dynamic Window

1. Fixed Window

If a fixed window of length k traverses a datastructre from start to finish it is guaranteed to traverse very component of length k in that.

```python
#Generic Template Code for Fixed Window Sliding Window Problem

left = right = 0
while right<n:
    #If the current window size has reached the required size of k, process the window
    if right-left+1 == k:
        process_window()
        left += 1
    right += 1
```

2. Dynamic Window

Problems that ask to find the longest or shortest subcomponent that satisfies a given condition.

When finding the longest subcomponent, the dynamics of expanding and shrinking are usually as follows.
- If the given window satisfies the condition, expand it to see if we can find longer window for the same.
- If the condition is violated then shrink the window until the condition is met again.

```python
#Generic Template Code for Dynamic Window Sliding Window Problem

left = right = 0
while right<n:
    #While the condition is violated, move the left pointer
    while condition_violated():
        left += 1
    # Once the condition is valid, process the window and move the right pointer to expand the window
    process_window()
    right += 1
```
